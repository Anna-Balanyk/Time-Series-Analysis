```python
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date


%matplotlib inline
```

```python
df = pd.read_csv('/kaggle/input/seattle-fremont-bridge-hourly-bicycle-counts/fremont-bridge-hourly-bicycle-counts-by-month-october-2012-to-present.csv')
df.head(10)
```

```python
df.columns = ['Date','Total', 'East', 'West']
df.info()
df.dropna()
```

```python
df['Datetime'] = pd.to_datetime(df['Date'])
df['Hour'] = df['Datetime'].dt.hour
df['Day'] = df['Datetime'].dt.day
df['Dayname'] = df['Datetime'].dt.day_name()
df['Week'] = df['Datetime'].dt.isocalendar().week
df['Month'] = df['Datetime'].dt.month
df['Year'] = df['Datetime'].dt.year
df
```

```python
df.groupby('Hour').mean()[['West', 'East']].plot()
```

```python
df.groupby('Dayname').sum()[['West', 'East']].plot()
```

```python
df.groupby('Month').mean()[['West', 'East']].plot()
```

```python
df.groupby('Week').mean()[['West', 'East']].plot()
```

```python
df.groupby('Year').mean()[['West', 'East']].plot()
```

```python
dayname_and_hour_data = df[['West','Dayname']][df['Hour']==12]

order_day_in_week = [
                     'Monday',
                     'Tuesday',
                     'Wednesday',
                     'Thursday',
                     'Friday',
                     'Saturday',
                     'Sunday',
                     ]

plt.figure(figsize=(10,4))
sns.stripplot(x='Dayname', y='West', data=dayname_and_hour_data, order=order_day_in_week )
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(x='Dayname', y='West', data=dayname_and_hour_data, order=order_day_in_week )
```
